```python
def add(a,b):
    c = a+b
    return c
def sub(a,b):
    c = a-b
    return c
def multi(a,b):
    c = a*b
    return c
def divi(a,b):
    c = a/b
    return c

while True:
    print("Please chose an option to start calculator")
    z = int(input("""1 for Addition
    2 for subtraction
    3 for multiplication 
    4 for Division
    5 for exit"""))
    
    if z == 5:
        break
    elif z == 1:
        a = int(input("Enter first number"))
        b = int(input("Enter the second number"))
        print(add(a,b))
    elif z == 2:
        a = int(input("Please enter greater number to avoid to get output below 0"))
        b = int(input("Enter the second number"))
        print(sub(a,b))
    elif z == 3:
        a = int(input("Enter first number"))
        b = int(input("Enter second number"))
        print(multi(a,b))
    elif z == 4:
        a = int(input("Enter Greater number"))
        b = int(input("Enter second number"))
        print(divi(a,b))
    else:
        print("please enter from the given options")
```
